```python
import pandas as pd
import numpy as np
import os
```

```python
from openpyxl import load_workbook, Workbook

def df_to_excel(df: pd.DataFrame, filename: str, sheet_name: str):
    # Otwórz istniejący plik Excel
    if os.path.exists(filename):
        workbook = load_workbook(filename)
    else:
        workbook = Workbook()

    # Konwertuj DataFrame na arkusz Excel
    writer = pd.ExcelWriter(filename, engine='openpyxl') 
    writer.book = workbook

    # Dopisz DataFrame do nowego arkusza
    df.to_excel(writer, sheet_name=sheet_name)

    # Zapisz zmiany w pliku
    writer.save()
```

```python
licznik=list(range(2,11))
arkusze=[y + str(x) for y in ['duf','duw'] for x in licznik]
arkusze
```

Unsupervised Length Average

```python
kolumna=('length','avg')
new_index_column=('number of attributes','Unnamed: 1_level_1')

#FEMALE
df2=pd.DataFrame(index=range(24,0,-1))
for ark in arkusze:
    df=pd.read_excel('Results2/Results_Female.xlsx',ark,header=[0,1],skiprows=[2],)
    df[[new_index_column,kolumna]]
    df=df.set_index(new_index_column)
    #ze względu na podwójne obliczenie dla pełnego zbioru:
    try:
        df2[ark]=df[kolumna] 
    except:
        df2[ark]=df[kolumna][1:]
display(df2)
df_to_excel(df2,'Results2/Results.xlsx',' unsup_Fe_len_avg')

#MALE
df2=pd.DataFrame(index=range(24,0,-1))
for ark in arkusze:
    df=pd.read_excel('Results2/Results_Male.xlsx',ark,header=[0,1],skiprows=[2],)
    df[[new_index_column,kolumna]]
    df=df.set_index(new_index_column)
    #ze względu na podwójne obliczenie dla pełnego zbioru:
    try:
        df2[ark]=df[kolumna] 
    except:
        df2[ark]=df[kolumna][1:]
display(df2)
df_to_excel(df2,'Results2/Results.xlsx',' unsup_Ma_len_avg')
```

```python
NoR=('number of rules','Unnamed: 8_level_1')
Acc=('mean','Unnamed: 11_level_1')
NoA=('number of attributes','Unnamed: 1_level_1')

#FEMALE
df2=pd.DataFrame(columns=['Discret method',
                          'Pos',
                          'avgL',
                          'avgS',
                          'rules',
                          'max_rules',
                          'Cl.acc[%]'])

for ark in arkusze:
    
    df=pd.read_excel('Results2/Results_Female.xlsx',ark,header=[0,1],skiprows=[2],)
    max_rules=max(df[NoR])

    df_max_mean=df.loc[df[Acc]==max(df[Acc])]
    df_min_NoR=df_max_mean.loc[df_max_mean[NoR]==max(df_max_mean[NoR])]
    df_min_NoR
    #index znalezionego rozwiązania
    idx = df_min_NoR.index[0]

    df2.loc[len(df2)] = [ark,
                        df_min_NoR[NoA][idx],
                        df_min_NoR[('length','avg')][idx], 
                        df_min_NoR[('support','avg')][idx], 
                        df_min_NoR[NoR][idx], 
                        max_rules, 
                        df_min_NoR[Acc][idx]]

display(df2)
df_to_excel(df2,'Results2/Results.xlsx',' Fe_Best')

#MALE
df2=pd.DataFrame(columns=['Discret method',
                          'Pos',
                          'avgL',
                          'avgS',
                          'rules',
                          'max_rules',
                          'Cl.acc[%]'])

for ark in arkusze:
    df=pd.read_excel('Results2/Results_Male.xlsx',ark,header=[0,1],skiprows=[2],)
    max_rules=max(df[NoR])

    df_max_mean=df.loc[df[Acc]==max(df[Acc])]
    df_min_NoR=df_max_mean.loc[df_max_mean[NoR]==max(df_max_mean[NoR])]
    df_min_NoR
    #index znalezionego rozwiązania
    idx = df_min_NoR.index[0]

    df2.loc[len(df2)] = [ark,
                        df_min_NoR[NoA][idx],
                        df_min_NoR[('length','avg')][idx], 
                        df_min_NoR[('support','avg')][idx], 
                        df_min_NoR[NoR][idx], 
                        max_rules, 
                        df_min_NoR[Acc][idx]]

display(df2)
df_to_excel(df2,'Results2/Results.xlsx',' Ma_Best')
```
